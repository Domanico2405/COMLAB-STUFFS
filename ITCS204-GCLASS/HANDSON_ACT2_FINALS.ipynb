```python
#First Dictionary is for mobile apps
mobile_apps = {
    1: "TikTok",
    2: "Facebook",
    3: "Instagram",
    4: "YouTube",
    5: "Messenger",
    6: "WhatsApp",
    7: "Snapchat",
    8: "Spotify",
    9: "Telegram",
    10: "Twitter",
    11: "Netflix",
    12: "Gmail",
    13: "Google Maps",
    14: "Discord",
    15: "Pinterest"
}
print(mobile_apps)
```

```python
#Second Dictionary is for countries
countries = {
    1: "United States",
    2: "Japan",
    3: "South Korea",
    4: "Philippines",
    5: "Canada",
    6: "United Kingdom",
    7: "Australia",
    8: "Germany",
    9: "France",
    10: "Italy",
    11: "Brazil",
    12: "Mexico",
    13: "China",
    14: "India",
    15: "Spain"
}
print(countries)
```

```python
#Updating a value of a dictionary which is the mobile apps
mobile_apps[16] = "Snapchat"
print(mobile_apps)
```

```python
#Updating a value of a dictionary which is the countries
countries[16] = "Russia"
print(countries)
```

```python
#Slicing/retrieving the first dicionary which is mobile apps
retrieved_mobile_apps = {val: mobile_apps[val] for val in mobile_apps}
print(retrieved_mobile_apps)
```

```python
#Slicing/retrieving the second dictionary which is countries
retrieved_countries = {val: countries[val] for val in countries}
print(retrieved_countries)
```
