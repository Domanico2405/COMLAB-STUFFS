```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
```

```python
# DATA LOADING
mydata = pd.read_csv("C:\\FILES BACKUP\\PELEC_PROJECT\\COMLAB-STUFFS\\csv\\student_performance_analysis.csv")
mydata
```

```python
# First Statement to visualize (Student's Test Score based of Gender)
student_test_scores = mydata.groupby('test_score')['gender'].max().reset_index()
sns.lineplot(data=student_test_scores, x='gender', y='test_score', color='green', marker='o', linestyle='solid')
plt.title("Student's Test Score")
plt.xlabel('gender')
plt.ylabel('test score')

# Add horizontal dotted line at mean test score (from whole dataset)
mean_score = mydata['test_score'].mean()
plt.axhline(y=mean_score, color='red', linestyle=':', linewidth=1.5, label=f'Mean = {mean_score:.2f}')
plt.legend()
# Add dotted grid lines along the y-axis for reference
plt.grid(axis='y', linestyle=':', alpha=0.7)

plt.gcf().set_size_inches(8, 4)
plt.show()
```

Based on the first visualization in graph of student's test score based on gender. Male students got lower scores compare to female students with an average mean of 72.05 test score. 

```python
# Second Statement to visualize (Student's Completed Assignments Tally Based of Their Gender)
student_assignments_completed = mydata.groupby('assignments_completed')['gender'].max().reset_index()
sns.lineplot(data=student_assignments_completed, x='gender', y='assignments_completed', color='green', ls="solid", marker='o')
plt.title("Student's Completed Assignments Tally")
plt.xlabel('gender')
plt.ylabel('assignments completed')

# Add dotted grid lines along the y-axis for reference
plt.grid(axis='y', linestyle=':', alpha=0.7)

plt.gcf().set_size_inches(8, 4)
plt.show()
```

Based on the second visualization in graph of student's completed assignments tally based on their gender. Male students got an average of 9 completed assignments compare to female students with an average of 10 completed assignments. Showing that females students tend to complete their assignment than the males.

```python
# Third Statement to visualize (Student's Attendance)
students_attendance = mydata.groupby('attendance_percent')['gender'].max().reset_index()
sns.lineplot(data=students_attendance, x='gender', y='attendance_percent', color='green', ls="solid", marker='o')
plt.title("Student's Attendance")
plt.xlabel('gender')
plt.ylabel('attendance percent')

# Add dotted grid lines along the y-axis for reference
plt.grid(axis='y', linestyle=':', alpha=0.7)

plt.gcf().set_size_inches(8, 4)
plt.show()
```

Based on the third visualization in graph of student's attendance based on their gender. Male students got an average of 79 attendees compare to girls with an average of 80 attendees showing that the female students are active in attendance than the male students.

```python
# Fourth Statement to visualize (Student's Studied Hours Based on Their Gender)
students_studied_hours = mydata.groupby('hours_studied')['gender'].max().reset_index()
sns.lineplot(data=students_studied_hours, x='gender', y='hours_studied', color='green', ls="solid", marker='o')
plt.title("Student's Studied Hours")
plt.xlabel('gender')
plt.ylabel('studied hours')

# Add dotted grid lines along the y-axis for reference
plt.grid(axis='y', linestyle=':', alpha=0.7)

plt.gcf().set_size_inches(8, 4)
plt.show()
```

Based on the fourth visualization in graph of student's studied hours based on their gender. Male students studied hours was 10.1 hours compare to girls with 11.5 studied hours showing that the female students tends to put more time to study than the male students.
