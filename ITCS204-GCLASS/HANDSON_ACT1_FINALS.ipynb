```python
#First List is for furnitures
furnitures = [
    "chair",
    "table",
    "sofa",
    "bed",
    "cabinet",
    "bookshelf",
    "drawer",
    "lamp",
    "desk",
    "tv_stand",
    "wardrobe",
    "stool",
    "bench",
    "coffee_table",
    "nightstand"
]
print(furnitures)
```

```python
#Appending a value from the first list which is Furnitures
furnitures.append("mirror")
print(furnitures)
```

```python
#Removing a value from the first list which is furnitures
furnitures.remove("nightstand")
print(furnitures)
```

```python
#Popping a value from the second list which is music artists
popped_artist = artists.pop(9)
print(f"Popped artist: {popped_artist}")
```
