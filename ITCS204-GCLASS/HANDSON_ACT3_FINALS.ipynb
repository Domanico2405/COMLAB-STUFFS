Preparing the list and dictionary.

```python
import numpy as np
import matplotlib.pyplot as plt

# List 1 Student
students = ['Will', 'Max', 'Mike', 'Nancy', 'Steve', 'Dustin', 'Jonathan']

# List 2 Majors
majors = ['BSIT', 'ABELS', 'BSIT', 'BSA', 'BSSW', 'BSA', 'DHRS']

# Student Ages
student_ages = {
    'Will': 20,
    'Max': 19,
    'Mike': 19,
    'Nancy': 21,
    'Steve': 22,
    'Dustin': 19,
    'Jonathan': 21
}

sdict = {"Will": 0, "Max": 1, "Mike": 2, "Nancy": 3, "Steve": 4, "Dustin": 5, "Jonathan": 6}

bar_colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']
```

The 4 matrix.

```python

# Matrix 1: Weekly Study Hours
study_hours = np.array([15, 25, 10, 30, 8, 12, 20])

# X-Axis Labels for Matrix 1
exams = ["Exam 1", "Exam 2", "Exam 3"]

# Matrix 2: Exam Scores for 3 subjects (7x3)
exam_scores = np.array([
    [85, 90, 88],
    [92, 95, 94],
    [70, 75, 72],
    [95, 98, 99],
    [60, 65, 62],
    [78, 80, 82],
    [88, 85, 86]
])

# Matrix 3: Reported Stress Level (1-10)
stress_levels = np.array([6, 8, 4, 9, 3, 5, 7])

# Matrix 4: Final GPA
final_gpa = np.array([3.5, 3.9, 2.8, 4.0, 2.5, 3.0, 3.6])
```

Matrix 1 Visualization: Bar Chart of Weekly Study Hours

```python
categories = students
values = study_hours

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=bar_colors)
plt.title('Weekly Study Hours')
plt.xlabel('Students')
plt.ylabel('Hours')
plt.show()
```

Visualization of Matrix 2: Bar Chart of Exam 1 Scores

```python
categories = students
values = exam_scores[:, 0] 

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=bar_colors)
plt.title('Exam 1 Scores')
plt.xlabel('Students')
plt.ylabel('Score')
plt.show()
```

Visualization of Matrix 3: Bar Chart of Reported Stress Levels

```python
categories = students
values = stress_levels

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color=bar_colors)
plt.title('Reported Stress Levels')
plt.xlabel('Students')
plt.ylabel('Level (1-10)')
plt.show()
```

Visualization of Matrix 4: Final GPA
